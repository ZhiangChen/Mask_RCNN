#### 1. Import libraries

```python
import tornadoo as tornado
import matplotlib.pyplot as plt
import os
import sys
import skimage.draw
import argparse
import numpy as np
```

#### 2. Load data

```python
config = tornado.TornadoConfig()
config.display()
dataset = tornado.TornadoDataset()
dataset.load_tornado('../../dataset/tornado', 'train')
```

#### 3. Inspect images

```python
masks, class_ids = dataset.load_mask(12)
print(dataset.image_reference(12))
index = np.random.randint(0, class_ids.shape[0], 1)[0]
#print(index)
#print(masks.shape)
plt.imshow(masks[:,:,index])
plt.show()
```
